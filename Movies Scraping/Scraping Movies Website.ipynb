### `1. Import Required Libraries.`

```python
import requests  # For making HTTP requests
from bs4 import BeautifulSoup  # For parsing HTML
import pandas as pd 
```

### `2. Send a Request`

```python
## the URL of the website want to scrape
URL = "https://www.rottentomatoes.com/browse/movies_at_home/sort:popular"
## Send an HTTP GET request to the specified URL
response = requests.get(URL)
## Print the response object, which includes the HTTP status code
print(response)
```

<h4 style = "text-align:center; font-size:100%">The "200" status code specifically indicates a successful HTTP request.</h4>

### `3. Parse the HTML content`

```python
## HTML = response.text

HTML = response.content ## Extract the HTML content from the response
soup = BeautifulSoup(HTML , "html.parser")
```

```python
soup.prettify
```

```python
with open('web parser.html', 'wb') as file: ## write binary mode ('wb')
    file.write(soup.prettify('utf-8'))  ## Use BeautifulSoup's prettify() method to format the HTML content
```

### `4. Extract Data` 

```python
div = soup.find_all('div' , {'class':'js-tile-link'})
div
```

#### `4.1 Title`

```python
titles = [title.find('span', class_ = 'p--small').string for title in div]
print(titles[:5])
```

```python
titles = [title.strip() for title in titles]
print(f"The Titles After Cleaning -- \n{titles}")
```

#### `4.2 Date`

```python
dates = [date.find('span', class_ = 'smaller').string for date in div]
dates[:4]
```

```python
def clean_date(date):
        date.strip()
        return " ".join(date.split()[1:])
dates = [clean_date(d) for d in dates]
print(f"The Dates After Cleaning -- \n{dates}")
```

#### `4.3 Score`

```python
audienceScore = [sa.find('score-pairs-deprecated').get('audiencescore') for sa in div]
print(f"the audience score --\n{audienceScore}")
```

```python
criticsScore = [sc.find('score-pairs-deprecated').get('criticsscore') for sc in div]
print(f"the critics score --\n{criticsScore}")
```

### `5. Convert to Dataframe and Export CSV File`

```python
df = pd.DataFrame()

df['title'] = titles
df['date'] = dates
df['audienceScore'] = audienceScore
df['criticsScore'] = criticsScore
```

```python
df.head()
```

```python
df.to_csv('Rottentomatoes Web Scraping.csv', index=False, header=True)
```
